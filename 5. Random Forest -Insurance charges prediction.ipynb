```python
#INSURANCE CHARGES PREDICTION USING RANDOM FOREST REGRESSOR

# Result: High score
# n_estimators=100,random_state=0,criterion=friedman_mse,max_features='sqrt'),    r_score = 0.8709457661833552

import pandas as pd
```

```python
# DATASET COLLECTION

dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python

dataset.columns
```

```python
# INPUT & OUTPUT SPLIT UP
# INDEPENDENT=INPUT
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
# DEPENDENT= OUTPUT
dependent=dataset[['charges']]
```

```python
dependent
```

```python

# TRAINING AND TEST SET SPLIT UP

from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test= train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
# Learning Phase
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100,random_state=0,criterion='friedman_mse',max_features='sqrt')
regressor.fit(X_train,Y_train)
```

```python
Y_pred=regressor.predict(X_test)
```

```python
#Evaluation

from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
r_score
```
